```python
import sqlite3

import pandas as pd

connection = sqlite3. connect('books.db')

pd.options.display.max_columns = 10
#Selecting the whole authors table
pd.read_sql('select * from authors', connection)
```

```python
#Selecting the whole titles table
pd.read_sql('select * from titles', connection)
```

```python
#Selecting the whole author isbn table
pd.read_sql('select * from author_ISBN', connection)
```

```python
#Selecting data from the titles table where the copyright of the book is after 2016
pd.read_sql("""Select title, edition, copyright from titles where copyright > 2016""", connection)
```

```python
#Selecting authors from the authors table where the last name starts with "D"
pd.read_sql("""select id, first, last from authors where last like 'D%'""", connection, index_col=['id'])
```

```python
#Selecting the author from the authors table where the first name starts with any letter followed by a "b"
pd.read_sql("""select id, first, last from authors where first like '_b%'""", connection, index_col=['id'])
```

```python
#Using the order by clause
pd.read_sql('select title from titles order by title', connection)
```

```python
#sorting by multiple columns
pd.read_sql("""select id, first, last from authors order by last, first""", connection, index_col=['id'])
```

```python
#Order by last name descending and then by the first name
pd.read_sql("""select id, first, last from authors order by last desc, first""", connection, index_col=['id'])
```

```python
#combining where and order by clause
pd.read_sql("""select isbn, title, edition, copyright from titles where title like '%How to Program' order by title""", connection)
```
